```python
# Filtrar las columnas que no contienen 'float'
columnas_no_float = [col for col in df.columns if 'float' not in str(df[col].dtype)]

# Limitar a columnas con un número manejable de valores únicos (ejemplo: menos de 50 valores únicos)
columnas_limitadas = [col for col in columnas_no_float if df[col].nunique() < 50]

# Iterar solo sobre las columnas filtradas y limitadas
for column in columnas_limitadas:
    conteo_valores = df[column].value_counts()
    valor_max = conteo_valores.max()

    # Graficar todos los valores
    plt.figure(figsize=(10, 6))
    ax = conteo_valores.plot(kind='bar')
    plt.xlabel('Valores')
    plt.ylabel('Frecuencia')

    # Título de la gráfica
    plt.title(f'Distribución de valores en "{column}"')

    # Rotación y estilo de los ejes
    plt.xticks(rotation=45, fontsize=8)
    plt.grid(axis='y', linestyle='--')
    plt.tight_layout()

    # Condición para aplicar la escala logarítmica
    if column not in ['Supervisor District']:
        ax.set_yscale('log')
    print(column)

    # Mostrar la gráfica
    plt.show()
```

#### b.1) Sustituir NaNs vacios por strings vacios para no tener conflictos de tipos de datos en columnas 'Voluntary Soft-Story Retrofit' y 'TIDF Compliance'

```python
for i in range (df['Supervisor District'].size):
    if isinstance(df['Supervisor District'][i], float):
        if np.isnan(df['Supervisor District'][i]):
            df['Supervisor District'][i] = ''
        else:
            df['Supervisor District'][i] = str(df['Supervisor District'][i])
    elif isinstance(df['Supervisor District'][i], str): 
        if df['Supervisor District'][i] == 'quince':
            df['Supervisor District'][i] = '15.0'
        elif df['Supervisor District'][i] == 'veinte':
            df['Supervisor District'][i] = '20.0'
        elif df['Supervisor District'][i] == 'diez':
            df['Supervisor District'][i] = '10.0'
```

### b) Investigar ventajas y posibles usos de esta tarea.

**Normalización:**
Es el proceso de transformación de los datos que los alinea con respecto a una escala común. Esto puede ser tanto para datos numéricos como para escalas de tiempo para comparar períodos similares

**Ventajas de normalizar los datos:**
La normalizacion de los datos hace que las magnitudes grandes no opaquen a las más chicas. Esto permite la comparación de los datos sin una subjetividad relativa a la naturaleza de los mismos. Por ejemplo, analizando datos de salud usando peso e índice de masa corporal, el peso tiene valores lejanos a cero y más elevados que el índice de masa corporal. Normalizar el peso podría ser tomar el mínimo y el máximo, de forma de disminuir el sesgo.
