#### b.3) Corregir el nombre de las calles, para obtener la primer letra en mayúscula y el resto el minúscula. 
En caso de que el primer caracter sea un número, todas las letras van en minúscula.

```python
def corregir_street_name(calle):
    if calle[0].isdigit():
        palabras = calle.lower().split()
        return palabras[0] + ' ' + ' '.join([palabra.capitalize() for palabra in palabras[1:]])
    else:
        return calle.title()

# Aplicar la función a toda la columna 'Street Name'
df['Street Name'] = df['Street Name'].apply(corregir_street_name)

df['Street Name']
```

## 5) Normalización de datos 

### a) Normalizar las variables continuas (numéricas)

Para normalizar las variables numéricas, se encontró que existen dos métodos utilizando pandas: Min-Max Scaling y Z-Score Normalization. Decidimos enfocarnos en el segundo, ya que el Z-Score Normalization (también conocido como estandarización) transforma los datos de manera que la media de los datos se convierte en 0 y la desviación estándar se convierte en 1, permitiendo que los datos sean comparables independientemente de la escala original. Aunque el Min-Max Scaling es útil para ajustar los datos a un rango específico, preferimos Z-Score ya que maneja mejor los valores atípicos y facilita la comparación entre variables.

```python
variables_num = df.select_dtypes(include=['float64', 'int64']).columns
df[variables_num] = (df[variables_num] - df[variables_num].mean()) / df[variables_num].std()
print(df.head())
```

### b) Investigar ventajas y posibles usos de esta tarea.

**Normalización:**
Es el proceso de transformación de los datos que los alinea con respecto a una escala común. Esto puede ser tanto para datos numéricos como para escalas de tiempo para comparar períodos similares

**Ventajas de normalizar los datos:**
La normalizacion de los datos hace que las magnitudes grandes no opaquen a las más chicas. Esto permite la comparación de los datos sin una subjetividad relativa a la naturaleza de los mismos. Por ejemplo, analizando datos de salud usando peso e índice de masa corporal, el peso tiene valores lejanos a cero y más elevados que el índice de masa corporal. Normalizar el peso podría ser tomar el mínimo y el máximo, de forma de disminuir el sesgo.
