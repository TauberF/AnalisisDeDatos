#### b.3) Corregir el nombre de las calles, para obtener la primer letra en may√∫scula y el resto el min√∫scula. 
En caso de que el primer caracter sea un n√∫mero, todas las letras van en min√∫scula.

```python
def corregir_street_name(calle):
    if calle[0].isdigit():
        palabras = calle.lower().split()
        return palabras[0] + ' ' + ' '.join([palabra.capitalize() for palabra in palabras[1:]])
    else:
        return calle.title()

# Aplicar la funci√≥n a toda la columna 'Street Name'
df['Street Name'] = df['Street Name'].apply(corregir_street_name)

df['Street Name']
```

## 5) Normalizaci√≥n de datos 

### a) Normalizar las variables continuas (num√©ricas)

Para normalizar las variables num√©ricas, se encontr√≥ que existen dos m√©todos utilizando pandas: Min-Max Scaling y Z-Score Normalization. Decidimos enfocarnos en el segundo, ya que el Z-Score Normalization (tambi√©n conocido como estandarizaci√≥n) transforma los datos de manera que la media de los datos se convierte en 0 y la desviaci√≥n est√°ndar se convierte en 1, permitiendo que los datos sean comparables independientemente de la escala original. Aunque el Min-Max Scaling es √∫til para ajustar los datos a un rango espec√≠fico, preferimos Z-Score ya que maneja mejor los valores at√≠picos y facilita la comparaci√≥n entre variables.

```python
variables_num = df.select_dtypes(include=['float64', 'int64']).columns
df[variables_num] = (df[variables_num] - df[variables_num].mean()) / df[variables_num].std()
print(df.head())
```

### b) Investigar ventajas y posibles usos de esta tarea.

**Normalizaci√≥n:**
Es el proceso de transformaci√≥n de los datos que los alinea con respecto a una escala com√∫n. Esto puede ser tanto para datos num√©ricos como para escalas de tiempo para comparar per√≠odos similares

**Ventajas de normalizar los datos:**
La normalizacion de los datos hace que las magnitudes grandes no opaquen a las m√°s chicas. Esto permite la comparaci√≥n de los datos sin una subjetividad relativa a la naturaleza de los mismos. Por ejemplo, analizando datos de salud usando peso e √≠ndice de masa corporal, el peso tiene valores lejanos a cero y m√°s elevados que el √≠ndice de masa corporal. Normalizar el peso podr√≠a ser tomar el m√≠nimo y el m√°ximo, de forma de disminuir el sesgo.
